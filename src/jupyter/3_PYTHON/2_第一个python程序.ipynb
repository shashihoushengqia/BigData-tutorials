## Python版本选择

### 📌 重要提示

Python目前有两个主要版本：
- **Python 2.x**：旧版本，已停止维护
- **Python 3.x**：最新版本，持续更新

> ⚠️ **本教程基于最新的Python 3.x版本**，请确保安装Python 3.x以获得最佳学习体验。

这两个版本**不兼容**，因此版本选择至关重要。

## 各平台安装指南

### 🪟 Windows系统

**方法一：官方安装程序（推荐）**

1. 访问 [Python官方网站](https://www.python.org/downloads/)
2. 下载 **Windows installer (64-bit)**
3. 运行 `python-3.x-amd64.exe` 安装包
4. **关键步骤**：务必勾选 **"Add Python 3.x to PATH"**
5. 点击 **"Install Now"** 完成安装

**方法二：使用包管理器Scoop**

```powershell
# 在PowerShell中执行
scoop install python
```

### 🍏 macOS系统

**方法一：官方安装程序**

1. 从Python官网下载macOS版安装程序
2. 双击运行并完成安装

**方法二：使用Homebrew**

```bash
# 在终端中执行
brew install python3
```

> 💡 注意：macOS系统自带Python 2.x，但我们需要安装Python 3.x

### 🐧 Linux系统

大多数Linux发行版都预装了Python，但可能需要手动安装Python 3：

```bash
# Ubuntu/Debian
sudo apt update
sudo apt install python3

# CentOS/RHEL
sudo yum install python3
```

## 验证安装

### 🔍 检查Python版本

安装完成后，在命令行中验证安装是否成功：

| 操作系统 | 命令 |
|----------|------|
| Windows | `python --version` |
| macOS/Linux | `python3 --version` |

预期输出示例：
```
Python 3.9.0
```

### 🖥️ 启动Python交互环境

在命令行中输入相应命令启动Python：

- **Windows**: `python`
- **macOS/Linux**: `python3`

成功启动后会看到类似以下的提示符：
```
>>> 
```

此时可以输入Python代码进行测试：
```python
>>> print("Hello, Python!")
Hello, Python!
```

## Python解释器类型

### 🏆 CPython（官方推荐）

- **最广泛使用的Python解释器**
- 用C语言开发，因此得名CPython
- 从Python官网下载的安装程序就是CPython
- **本教程所有代码均在CPython下测试**

### 🔧 其他解释器

| 解释器 | 特点 | 适用场景 |
|--------|------|----------|
| **IPython** | 增强的交互式解释器，提供更好的用户体验 | 数据分析、交互式开发 |
| **PyPy** | 使用JIT技术，执行速度更快 | 性能敏感的应用 |
| **Jython** | 运行在Java平台上，可编译为Java字节码 | Java生态系统集成 |
| **IronPython** | 运行在.NET平台上，可编译为.NET字节码 | .NET生态系统集成 |

> 💡 **建议**：初学者应使用CPython，它是标准且最稳定的Python解释器。

## 常见问题解决

### ❌ "python不是内部或外部命令"

**原因**：PATH环境变量未正确配置

**解决方案**：
1. 重新运行安装程序
2. 确保勾选 **"Add Python to PATH"** 选项
3. 或手动将Python安装路径添加到系统PATH中

### 🔄 Python 2 vs Python 3混淆

**现象**：在macOS/Linux上输入`python`启动的是Python 2

**解决方案**：始终使用`python3`命令来运行Python 3

### 📦 安装后无法导入模块

**原因**：可能安装了多个Python版本导致冲突

**解决方案**：
- 使用虚拟环境隔离不同项目
- 确认使用正确的pip版本（`pip3`对应Python 3）

> ✅ **最佳实践**：安装完成后立即测试基本功能，确保环境配置正确。
