## Python运行模式

### 🖥️ 命令行模式 vs Python交互模式

| 模式 | 特点 | 用途 |
|------|------|------|
| **命令行模式** | 系统命令行环境，提示符如 `PS C:\>` | 运行Python脚本文件、系统操作 |
| **Python交互模式** | Python解释器环境，提示符为 `>>>` | 快速测试代码、学习语法 |

### 🔁 模式切换

**进入Python交互模式：**
```bash
# Windows
python

# macOS/Linux
python3
```

**退出Python交互模式：**
```python
>>> exit()
```

> 💡 **提示**：交互模式适合快速测试，但代码无法保存；实际开发应使用文本编辑器编写脚本文件。

## 使用文本编辑器

### 🛠️ 推荐编辑器

**Visual Studio Code (VS Code)**
- 微软出品的轻量级代码编辑器
- 跨平台支持（Windows、macOS、Linux）
- 免费且功能强大
- 完美支持Python开发

### ⚠️ 避免使用的编辑器

- **Microsoft Word**：保存的不是纯文本文件
- **Windows记事本**：会在文件开头添加UTF-8 BOM特殊字符，导致程序运行错误

### 📝 创建第一个Python文件

1. 在VS Code中创建新文件
2. 输入以下代码：
```python
print('hello, world')
```
3. **注意**：`print`前面不要有任何空格
4. 保存文件为 `hello.py`（必须以`.py`结尾）
5. 文件名只能包含英文字母、数字和下划线

## 运行Python程序

### ▶️ 基本运行方法

1. 打开命令行窗口
2. 切换到Python文件所在目录：
```bash
cd C:\your\directory\path
```
3. 运行Python程序：
```bash
# Windows
python hello.py

# macOS/Linux
python3 hello.py
```

预期输出：
```
hello, world
```

### ❌ 常见错误

**文件不存在错误：**
```
python: can't open file 'hello.py': [Errno 2] No such file or directory
```

**解决方案：**
- 确认当前目录下存在该文件
- 使用`cd`命令切换到正确的目录

### 🚀 直接运行Python文件（macOS/Linux）

在文件第一行添加shebang注释：
```python
#!/usr/bin/env python3

print('hello, world')
```

赋予执行权限并运行：
```bash
chmod a+x hello.py
./hello.py
```

## 输入和输出

### 📤 输出函数 `print()`

**基本用法：**
```python
# 输出字符串
print('hello, world')

# 输出多个值（自动用空格分隔）
print('The quick brown fox', 'jumps over', 'the lazy dog')

# 输出数字和计算结果
print(300)
print(100 + 200)

# 混合输出字符串和计算结果
print('100 + 200 =', 100 + 200)
```

**输出结果：**
```
hello, world
The quick brown fox jumps over the lazy dog
300
300
100 + 200 = 300
```

### 📥 输入函数 `input()`

**基本用法：**
```python
# 获取用户输入
name = input()

# 显示输入内容
print(name)
```

**带提示的输入：**
```python
name = input('请输入您的姓名：')
print('hello,', name)
```

> 💡 **重要概念**：变量是用来存储数据的容器。在Python中，变量可以存储整数、浮点数、字符串等各种类型的数据。

````python
# 完整的第一个Python程序示例

# 获取用户输入
name = input('请输入您的姓名：')

# 输出个性化问候
print('hello,', name)

# 计算并输出一些有趣的数字
age = input('请输入您的年龄：')
print(f'您今年 {age} 岁，明年就是 {int(age) + 1} 岁了！')

# 简单的数学运算
print('让我们来做个简单的计算：')
print('100 + 200 =', 100 + 200)
print('10 * 5 =', 10 * 5)
````

## 最佳实践总结

### ✅ 开发流程

1. **选择合适的编辑器**：推荐VS Code
2. **创建.py文件**：确保文件扩展名为`.py`
3. **编写代码**：注意缩进和语法
4. **保存文件**：使用有意义的文件名
5. **运行测试**：在命令行中运行程序
6. **调试修正**：根据错误信息修正代码

### 📋 文件命名规范

- 只能包含英文字母、数字和下划线
- 不能以数字开头
- 必须以`.py`结尾
- 推荐使用小写字母和下划线组合

### 🎯 学习建议

- 从简单的程序开始，逐步增加复杂度
- 多练习输入输出操作
- 理解变量的概念和使用
- 养成良好的代码格式习惯
