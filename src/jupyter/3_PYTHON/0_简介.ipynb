## 什么是Python？

Python是一种**高级编程语言**，具有以下显著特点：

### 代码简洁性

完成同一个任务，不同语言的代码量差异很大：
- C语言：约1000行代码
- Java语言：约100行代码  
- **Python语言：约20行代码**

### 执行效率 vs 开发效率

| 语言 | 开发效率 | 执行速度 |
|------|----------|----------|
| C    | 低       | 快 (1秒) |
| Java | 中       | 中 (2秒) |
| Python | **高** | 慢 (10秒) |

> **注意**：代码少的代价是运行速度相对较慢，但对于大多数应用场景，这种权衡是值得的。

## Python能做什么？

### ✅ Python擅长的领域

- **日常任务自动化**：自动备份文件、批量处理数据等
- **Web开发**：YouTube、Instagram、Dropbox等知名网站
- **游戏后台开发**：许多在线游戏的服务器端
- **数据分析**：科学计算、统计分析
- **人工智能**：机器学习、深度学习
- **网络爬虫**：数据采集和信息提取

### ❌ Python不适合的领域

- **操作系统开发**：需要使用C语言
- **手机原生应用**：
  - iOS：Swift/Objective-C
  - Android：Java/Kotlin
- **高性能3D游戏**：最好使用C/C++
