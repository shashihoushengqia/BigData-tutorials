### 🚀 Git 极简入门教程：从零到上传代码

#### 第一步：准备工作

在开始之前，你需要两样东西：
1.  **Git 软件**：在你的电脑上安装 Git。
2.  **GitHub 账号**：在 GitHub 网站上注册一个账号。

**1. 安装 Git**
*   **Windows**: 去 [Git 官网](https://git-scm.com/downloads) 下载安装包，一路点击 "Next" 默认安装即可。
*   **Mac**: 打开终端，输入 `git --version`。如果提示安装，按提示安装即可；或者去官网下载。
*   **Linux**: 终端输入 `sudo apt install git` (Ubuntu/Debian)。

**2. 注册 GitHub**
*   访问 [GitHub](https://github.com/)，点击 Sign up 注册账号。

---

#### 第二步：配置你的身份（只需做一次）

安装好 Git 后，你需要告诉 Git 你是谁。打开你的终端（Terminal、CMD 或 PowerShell），依次输入以下两条命令（替换成你自己的名字和邮箱）：

```bash
git config --global user.name "你的名字"
git config --global user.email "你的邮箱"
```
*注意：这里的邮箱最好是你注册 GitHub 时使用的邮箱。*

---

#### 第三步：在 GitHub 上创建仓库

1.  登录 GitHub，点击右上角的 **+** 号，选择 **New repository**。
2.  **Repository name**：填写你的项目名字（例如 `my-first-project`）。
3.  **Public/Private**：选择 Public（公开）或 Private（私有）。
4.  **勾选 "Add a README file"**（这步可选，但建议勾选，方便初始化）。
5.  点击 **Create repository**。

创建成功后，你会看到一个页面，上面有一个网址，类似这样：
`https://github.com/你的用户名/my-first-project.git`
**复制这个地址，后面要用到。**

---

#### 第四步：上传本地代码

假设你的代码放在电脑的 `D:\code\my-project` 文件夹下。

1.  **进入文件夹**
    在终端中进入你的项目文件夹：
    *   Windows: `cd /d D:\code\my-project`
    *   Mac/Linux: `cd ~/code/my-project`

2.  **初始化仓库**
    在文件夹内输入：
    ```bash
    git init
    ```
    *这步会把这个文件夹变成一个 Git 可以管理的仓库。*

3.  **把文件添加到暂存区**
    输入：
    ```bash
    git add .
    ```
    *注意 `add` 后面有个空格和点 `.`，代表把文件夹里**所有**文件都加进去。*

4.  **提交文件**
    输入：
    ```bash
    git commit -m "第一次提交"
    ```
    *`-m` 后面的话是本次提交的备注，方便你以后看，内容可以自定义。*

5.  **关联远程仓库**
    输入（替换为你刚才复制的地址）：
    ```bash
    git remote add origin https://github.com/你的用户名/my-first-project.git
    ```

6.  **推送到 GitHub**
    输入：
    ```bash
    git push -u origin master
    ```
    *如果报错，可能是 GitHub 现在的主分支名改为了 `main`，如果报错提示找不到 master，请尝试把 `master` 换成 `main` 再试一次。*
    ```bash
    git push -u origin main
    ```

---

#### 💡 常见问题与小贴士

1.  **执行 `git push` 时要求输入账号密码？**
    *   现在的 GitHub 已经不支持密码推送了，推荐使用 **SSH Key** 或者 **Personal Access Token (PAT)**。
    *   **小白最简单的办法**：如果你使用 HTTPS 地址推送，GitHub 会弹窗让你在浏览器登录，登录一次后通常就会记住凭证。

2.  **每次都要输入密码怎么办？**
    *   Windows 用户如果安装了 Git Credential Manager，通常只需输入一次 Token 后就会自动保存。
    *   如果不想配置 SSH，可以在推送 URL 中嵌入用户名和 Token（不推荐长期这样做，但在本地环境很方便）：
        `git remote set-url origin https://你的Token@github.com/你的用户名/仓库名.git`

3.  **主分支名是 master 还是 main？**
    *   以前叫 `master`，现在为了更包容，GitHub 默认改叫 `main` 了。如果你不确定，在 GitHub 仓库页面的 "Code" 下拉菜单里看一眼默认分支名是什么，就用那个。
