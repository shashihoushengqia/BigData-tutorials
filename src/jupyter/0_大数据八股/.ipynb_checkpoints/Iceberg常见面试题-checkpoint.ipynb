### 为什么选择 Iceberg？它解决了什么问题？

首先我先说说相比于hive的优势
第一点就是iceberg支持原子性，Hive 的 INSERT OVERWRITE 不是原子的，下游可能看到部分数据。
第二点就是视图变更，我说一下传统hive的痛点吧，以前Hive 修改列名、类型或顺序非常困难且危险，例如要新增一个字段记录用户来源，或者修改某个字段的类型以容纳更大数值，传统方式通常需要重写整个表，对大数据量时，这意味这漫长的停机等待，
虽然可以通过一些手段，比如说通过临时表解决，但计算资源的消耗也不可忽视，导致修改结构的成本太高，团队不敢轻易调整模型，导致表结构里面存有大量无用字段。但是iceberg支持完整的视图变更
然后第三点是Iceberg支持隐藏分区，我又来说一下传统Hive分区的一些痛点，hive分区需要在表中每一行数据重复存储，这在数据量大时会消耗巨大的存储空间，然后是查询逻辑复杂容易出错，我们去查的时候需要知道底层的分区时按天还是按小时，
一旦查询条件与底层分区格式不完全匹配，就会导致全表扫描，再然后是演进很困难，比如以前按天的分区不适用了，要改成按小时，就必须重写整个文件，iceberg通过将分区信息存储在元数据文件中，解决了这些问题
然后再说说一些特别的功能，比如说iceberg支持时间旅行和多引擎支持等hive不具备的功能
当然虽然iceberg引入了很多优势，但也是有劣势的，它引入了额外的复杂度，如果优化不当可能会比hive更慢，效率更低

### 为什么不直接用Iceberg，ADS层为什么要选择Apache Doris

首先我讲一下我引入Apache Doris的原因哈，首先它支持高并发点查，Iceberg的设计理念是存算分离的，每次查询都需要从对象存储中读数据，会引入一些不必要的延迟，其次Iceberg的索引结构分3级，第一级是分区索引，第二级是清单元数据索引，第三级是底层格式parquert的页级别索引，分区定位到文件夹，清单定位到文件，页索引定位到行组，这些索引的级别都比较粗，无法精确定位到行，按照Iceberg的最佳实践你读一行数据可能就会扫描好几十MB的数据，还得解析清单文件，这个过程在高并发点查场景下引入的延迟就不可接受了，我再说说Apache Doris，它支持存算一体的模式，先天就屏蔽了对象存储延迟问题，然后支持倒排索引，虽然会引入比较大的存储开销，但它毕竟是行级别的索引，在高并发点查情况下可以做到比较低的延迟，再者它针对高并发点查场景做了短路径改写、行列混存等一系列优化手段，它先天就适合这个场景，最后就是Apache Doris的存储成本比较高，需要高性能的nvme硬盘才能支撑它的性能，但ADS层先天比较少的数据屏蔽了这个缺点。

### Iceberg 的快照机制是如何工作的？

Iceberg表每次写入时都会写入一个快照，它是不可变的，快照之间通过增量快照链关联，支持时间旅行查询，查询时通过快照 ID 或时间戳访问历史数据

### Iceberg 的隐藏分区是什么？

用户无需在查询中手动添加分区过滤函数，Iceberg 自动将分区转换函数应用于查询谓词，例如：按 hour(timestamp) 分区，查询 WHERE timestamp = '2024-01-01 10:00:00' 时，Iceberg 自动转换为小时级过滤，避免了 Hive 中常见的 WHERE hour_col = hour('2024-01-01 10:00:00') 手动转换错误

### Iceberg的表格式由哪些核心文件组成？

1. metadata文件，包含表的schema、分区规范、快照等信息。
2. 清单文件，记录数据文件的列表以及每个数据文件的分区、行数、文件大小等统计信息。
3. 数据文件，实际存储数据的文件，还分Position DELETE文件和等值删除文件，及向量删除文件
4. catalog版本文件，通常记录在rest catalog中，是乐观锁控制的关键

### Iceberg 的 Catalog 类型有哪些？

Hive Metastore Catalog：传统方案，兼容性好
REST Catalog：云原生推荐，解耦元数据存储
Hadoop Tables：无 Catalog 依赖，直接基于文件系统，但事务控制较弱
JDBC Catalog：简单易用，但生态不成熟，有单点瓶颈
Glue Catalog：国外生态，国内我不是很了解

### 说说Iceberg有哪些写入模式？

Iceberg的写入模式分两种，第一种是copy-on-write，我来举个例子，它的原理是通过在写入时就更新掉文件的内容，比如说你修改一条数据，它会找到之前存储这条数据的文件，读到内存中，修改文件内容重写一个文件，并生成新的清单标识以前的文件已经删除，这样做的好处是读取很快，只需要读取重写后的文件即可，存储也比较干净，劣势也很明显，写入很慢嘛，因为要重写一整个文件，按照最佳实践可能是128MB，256MB之类的，这在更新比较多的情况下写放大就比较严重了，第二种是merge-on-read，还是刚才那个例子，你更新一条数据，它直接写入一个删除文件，标识哪个文件的哪行已经被删除了，然后再写入一个新的数据文件，写入新的更新后的数据即可，然后在更新清单即可，好处是没有写放大，只需要写入很小的更新后数据即可，缺点是引入了读放大，下游读取时需要读取原文件删除文件和新文件，但是可以通过维护任务减少问题。都说到这了，我在唠下删除文件的格式吧，Iceberg删除文件分三种，第一种是位置删除，它直接记录了将要删除文件的行号，可以做到精确删除，然后是等值删除，它不会记录删除的行号，它直接就不知道要删除哪行，但它记录了哪些记录需要删除，等应用层比如spark在读取时在内存中删除匹配相应的记录来删除，这种模式写入是最快的，因为它相比位置删除少了一个找到文件数据行号的操作。然年后是Iecberg v3格式引入的删除向量，这是对位置删除的一个增强版，它不再data文件夹中生成一个删除文件，取而代之的是在清单文件中维护一个删除行的稀疏数组，再通过分段压缩，使其占用的空间极小，好处是只需要维护一个数组，不需要像位置删除一样，需要维护大量占用空间比较大的删除文件。

### Iceberg 如何保证 ACID 事务？

Iceberg是通过乐观并发控制的：Iceberg假设冲突很少发生。首先，它的写操作会生成新的元数据文件和清单文件，提交时，会尝试原子性地更新版本元数据文件，如果检测到当前版本已被其他事务修改，则重试或失败，同时每次提交都会生成一个新的快照，旧快照依然存在，保证了读操作不会因为写操作而报错。
这只是概念，我来举个实际的例子，iceberg在写入任务时会先从元数据库（可能时rest，可能是jdbc，可能是hadoop）读一个版本号，读在自己的内存中，然后开启写入任务，等写完以后会提交，提交的时候会再次检查元数据库的版本号，如果和自己内存的版本号对得上，那就写入成功，更新版本号，如果对不上了，说明被其它写入任务修改了，如果继续提交则会冲突，因此直接报错，不更新版本号，不提交快照版本，因此写入的数据不会被任何快照所引用，此时就生成了孤儿文件，可以通过提交维护任务来删除这些孤儿文件。

### 从你的项目中来看，数据写入任务10分钟就开始了，和维护任务遇到了冲突怎么办，不会失败吗？

会遇到冲突，但也可以通过配置来避免，方案一：为了保证计算任务的时效性，我们允许部分失败维护任务，开启了Iceberg的部分提交机制，它可以先维护部分文件，先提交了再说，直到遇到写入任务的冲突，导致部分失败，然后通过Airflow的重试机制，多维护几次，直到任务维护完成，方案二：如果业务时效性能容忍牺牲，可以为写入任务配置维护任务的依赖，直到维护任务完成再继续写入，但以上两种方案离不开Iceberg维护任务的优化，比如维护任务的时间窗口设置需要合理，大的维护任务可以适当提高并行度，分层次的维护，我这边分了4级，首先在计算引擎就要配置合理的参数，减少写入的文件数量，然后是维护任务L0（写入完成即极小范围的维护，下游不必依赖该任务，使其形成异步的原理），L1（两小时维护一次中等规模的维护，下游也不必依赖该任务，形成异步的原理），L2层次的维护（每天低峰期维护一次大规模的维护，选择业务低峰期，可异步可同步），实际上在要求更高的系统中，应该有一个服务程序实时监控Iceberg表状态判断Iceberg表是否需要维护了，比如说文件数量是不是多了，小文件是不是多了，一旦满足条件就开启合并，其实paimon在这方面表现更好，他的锁是bucket级别的，写入时只能写入一个bucket，每个bucket并行维度，即使并发冲突了，维护也在内部进行，不需要用户自己处理重试
