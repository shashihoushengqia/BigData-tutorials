#### 统计需求
1. 计算每个用户的购买行为指标（仅统计已完成状态）：
* 总订单数
* 总消费金额
* 平均订单金额
* 首次购买日期
* 最近购买日期
* 购买天数跨度（最近购买日期 - 首次购买日期）
查询结果按总消费金额、总订单数降序排序，相等则按用户id升序，金额类指标保留两位小数

#### 数据探索

```python
%%sql
-- 查询表结构
DESC dataspire_catalog.ecommerce.ods_orders_mi;
```

```python
%%sql
-- 简单检查表的粒度及数据结构，一般还需要配合业务系统一起观察，如果对表数据有疑问需要咨询相关开发
SELECT
    *
FROM
    dataspire_catalog.ecommerce.ods_orders_mi t
WHERE
    1 = 1
LIMIT 10
;
```

```python
%%sql
-- 查询表数据量
-- 能看到一个月大约有600w数据左右，属于比较小的数据量，且每个月的订单数差异不大，不涉及数据倾斜
SELECT
    SUBSTR(t.order_date, 1, 7) AS biz_month
  , COUNT(0)                   AS order_cnt
FROM
    dataspire_catalog.ecommerce.ods_orders_mi t
WHERE
      1 = 1
GROUP BY
    biz_month
;
```

```python
%%sql
-- 查询表数据量大小
-- 平均每5万行数据占用1mb数据，一个月约200MB数据左右，数据大小比较小
SELECT
    SUM(file_size_in_bytes) / 1024 / 1024 AS size_mb
FROM
    dataspire_catalog.ecommerce.ods_orders_mi.files
WHERE 1 = 1
;
```

```python
%%sql
-- 第一次查询，使用Spark SQL的默认设置，读取并行度依赖底层的iceberg文件数，一般优化比较好的iceberg会控制在一个非常合适的并行度，shuffle并行度依赖于spark.sql.shuffle.partitions，老言设置的全局配置为24。
-- 从下方的web ui中可以得出结论，SparkSQL规划了两个阶段，并行度分别为6和24，运行时间为2.6s
SELECT
    t.user_id                     AS user_id -- 用户id
  , COUNT(0)                      AS order_cnt -- 总订单数
  , ROUND(SUM(t.final_amount), 2) AS tol_amt -- 总消费金额
  , ROUND(IF(COUNT(0) = 0, 0, SUM(t.final_amount) / COUNT(0)),
          2)                      AS avg_amt -- 平均消费金额
  , MIN(t.order_date)             AS first_order_date -- 首次购买时间
  , MAX(t.order_date)             AS last_order_date -- 最近购买时间
  , DATEDIFF(MAX(t.order_date), MIN(t.order_date)) +
    1                             AS days_diff -- 购买天数跨度
FROM
    dataspire_catalog.ecommerce.ods_orders_mi t
WHERE
      1 = 1
  AND t.status = '已完成'
  AND t.order_date >= CAST(TO_TIMESTAMP(ADD_MONTHS(CURRENT_TIMESTAMP, -1)) AS STRING)
GROUP BY
    t.user_id
-- 按照总消费金额降序排序
ORDER BY
    tol_amt DESC, order_cnt DESC, user_id
-- 限制输出结果为10条，虽然题目无要求，但这是非常推荐的一个实验用的习惯，可以极大地节省网络带宽
LIMIT 10
;
```

```python
%%sql
-- 第二次查询，手动优化并行度，数据量比较小，没有必要使用24个shuffle并行度，3-5个即可了，这次手动调整shuffle并行度
-- /*+ COALESCE(numPartitions) */用于减少shuffle并行度，不会像/*+ REPARTITION(numPartitions) */还会再开一个shuffle阶段
SELECT
    /*+ COALESCE(3) */
    t.user_id                     AS user_id -- 用户id
  , COUNT(0)                      AS order_cnt -- 总订单数
  , ROUND(SUM(t.final_amount), 2) AS tol_amt -- 总消费金额
  , ROUND(IF(COUNT(0) = 0, 0, SUM(t.final_amount) / COUNT(0)),
          2)                      AS avg_amt -- 平均消费金额
  , MIN(t.order_date)             AS first_order_date -- 首次购买时间
  , MAX(t.order_date)             AS last_order_date -- 最近购买时间
  , DATEDIFF(MAX(t.order_date), MIN(t.order_date)) +
    1                             AS days_diff -- 购买天数跨度
FROM
    dataspire_catalog.ecommerce.ods_orders_mi t
WHERE
      1 = 1
  AND t.status = '已完成'
  AND t.order_date >= CAST(TO_TIMESTAMP(ADD_MONTHS(CURRENT_TIMESTAMP, -1)) AS STRING)
GROUP BY
    t.user_id
-- 按照总消费金额降序排序
ORDER BY
    tol_amt DESC, order_cnt DESC, user_id
-- 限制输出结果为10条，虽然题目无要求，但这是非常推荐的一个实验用的习惯，可以极大地节省网络带宽
LIMIT 10
;
```

```python
%%sql
-- 第三次查询，不再手动优化分区数，而是通过SparkSQL的AQE机制自动优化，详情请参考AQE章节
-- 能够看到AQE将一个job拆成了两个job，并在一个job之后收集了动态统计信息将shuffle并行度修改成了1，但总运行时间由1.5s增加到了2.5s，这是因为在小规模且简单的统计查询中做规划引入了更多的延时，导致副作用，因此只建议复杂查询中打开
SELECT
    t.user_id                     AS user_id -- 用户id
  , COUNT(0)                      AS order_cnt -- 总订单数
  , ROUND(SUM(t.final_amount), 2) AS tol_amt -- 总消费金额
  , ROUND(IF(COUNT(0) = 0, 0, SUM(t.final_amount) / COUNT(0)),
          2)                      AS avg_amt -- 平均消费金额
  , MIN(t.order_date)             AS first_order_date -- 首次购买时间
  , MAX(t.order_date)             AS last_order_date -- 最近购买时间
  , DATEDIFF(MAX(t.order_date), MIN(t.order_date)) +
    1                             AS days_diff -- 购买天数跨度
FROM
    dataspire_catalog.ecommerce.ods_orders_mi t
WHERE
      1 = 1
  AND t.status = '已完成'
  AND t.order_date >= CAST(TO_TIMESTAMP(ADD_MONTHS(CURRENT_TIMESTAMP, -1)) AS STRING)
GROUP BY
    t.user_id
-- 按照总消费金额降序排序
ORDER BY
    tol_amt DESC, order_cnt DESC, user_id
-- 限制输出结果为10条，虽然题目无要求，但这是非常推荐的一个实验用的习惯，可以极大地节省网络带宽
LIMIT 10
;
```
